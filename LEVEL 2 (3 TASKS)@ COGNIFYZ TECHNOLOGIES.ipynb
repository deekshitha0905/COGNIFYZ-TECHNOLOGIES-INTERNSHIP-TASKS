ANALYZE THE AVAILABILITY OF ONLINE DELIVERY AMONG RESTAURANTS WITH DIFFERENT PRICE RANGES

```python
import pandas as pd
df=pd.read_csv("Dataset.csv")
totalrestaurants_by_price_range=df["Price range"].value_counts()
onlinedelivery_counts=df['Has Online delivery'].value_counts()
percentage_onlinedelivery_by_pricerange = {}
for pricerange in totalrestaurants_by_price_range.index:
      subset_df = df[df["Price range"] == pricerange]
      onlinedelivery_counts = subset_df['Has Online delivery'].value_counts()
      percentage_onlinedelivery = {onlinedelivery_counts.get('Yes',0)/totalrestaurants_by_price_range[pricerange]*100}
      percentage_onlinedelivery_by_pricerange[pricerange] = percentage_onlinedelivery

print("percentage of restaurants offering online delivery by price range:")
print(percentage_onlinedelivery_by_pricerange)
```

IDENTIFY THE COLOR THAT REPRESENTS THE HIGHEST AVERAGE RATING AMONG DIFFERENT PRICE RANGES

```python
average_rating_by_pricerange=df.groupby("Price range")['Aggregate rating'].mean()
highest_average_rating_by_pricerange=average_rating_by_pricerange.idxmax()
highest_rating_rows=df[df['Price range'] == highest_average_rating_by_pricerange]
highest_avereage_rating_color=highest_rating_rows['Rating color'].iloc[0]
print('color that represents the highest average range:',highest_avereage_rating_color)
```

                        TASK 3 FEATURE ENGINEERING

EXTRACT ADDITIONAL FEATURES FROM THE EXISTING COLUMNS, SUCH AS THE LENGTH OF THE RESTAURANT NAME OR ADDRESS

Extracting length:

```python
df['Restaurant Name Length'] = df['Restaurant Name'].apply(lambda x:len(x))
df['Address Length'] = df['Address'].apply(lambda x:len(x))
print(df)

```

List():

```python
df['Restaurant Name List'] = df['Restaurant Name'].apply(lambda x:list(x))
df['Address List'] = df['Address'].apply(lambda x:list(x))
print(df)
```

Enumerare():

```python
df['Restaurant Name List'] = df['Restaurant Name'].apply(lambda x:enumerate(x))
df['Address List'] = df['Address'].apply(lambda x:enumerate(x))
print(df)
```

Frozenset():

```python
df['Restaurant Name List'] = df['Restaurant Name'].apply(lambda x:frozenset(x))
df['Address List'] = df['Address'].apply(lambda x:frozenset(x))
print(df)
```

CREATE NEW FEATURES LIKE "HAS TABLE BOOKING" OR "HAS ONLINE DELIVERY" BY ENCODING CATEGORICAL VARIABLES

```python
df['Has Table booking']=df['Has Table booking'].map({'Yes':1,'No':0})
print(df)
```

```python
df['Has Online Delivery']=df['Has Online delivery'].map({'Yes':1,'No':0})
print(df)
```
